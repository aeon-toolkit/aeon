## Series

```python
from aeon.datasets import load_airline
from aeon.visualisation import plot_correlations, plot_lags, plot_series
```

```python
airline = load_airline()

plot_series(airline)
```

```python
airline_train, airline_test = airline[:-24], airline[-24:]
airline_mid = airline[60:96] + 100

plot_series(airline_train, airline_test, airline_mid)
```

```python
plot_lags(airline, lags=2)
```

```python
plot_lags(airline, lags=[1, 2, 3])
```

```python
plot_correlations(airline)
```
